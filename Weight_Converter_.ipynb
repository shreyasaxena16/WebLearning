```python
weight_pounds = float(input('What is your weight in pounds?'))
formula = 0.45359237
weight_kg = weight_pounds * formula
print(weight_kg)
```

```python
kg_weight = float(input('What is your weight in kilograms?'))
equation =  0.45359237
pounds_weight = kg_weight / equation
print(pounds_weight)
```

```python
unit = input('Pounds or Kilograms?').lower();
if unit == 'kilograms' or  unit == 'pounds':
  weight = float(input('What is your weight?'))
  number = 0.45359237
  if unit == 'pounds':
    lbs_weight = weight * number
    print(lbs_weight)
  elif unit == 'kilograms':
    kilo_weight = weight / number
    print(kilo_weight)


else:
   print('Wrong Unit='+unit)
```
