```python
year = float(input('What year?'))
if year % 4 == 0:
  year_two = year % 100
  if year_two == 0:
    year_three = year % 400
    if year_three == 0:
      print('This year is a leap year')
elif year % 4 == 0:
  year_two = year % 100
  if year_two != 0:
    print('This year is a leap year')
```

```python
yeary = float(input('What year?'))
if yeary % 4 == 0:
  year_2 = yeary % 100
else:
  print('This year is not a leap year')

if year_2 == 0:
  year_3 = yeary % 400
else:
  print('This year is a leap year')

if year_3 == 0:
  print('This year is a leap year')
else:
  print('This year is not a leap year')
```
