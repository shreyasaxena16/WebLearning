```python
number = float(input('Put a number'))
number_2 = number // 2
number_3 = number_2 * 2
output = number - number_3
if output == 0:
  print('even')
else:
  print('odd')
```

```python
num= float(input('Number'))
divisor = float(input('Divided by?'))
num_two = num // divisor
num_three = num_two * divisor
remainder = num - num_three
if remainder == 0:
  print (f"{num} is divisible by {divisor}")
else:
  print(f"{num} is not divisible by {divisor}")
```
